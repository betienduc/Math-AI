## Hàm render Latex ___(hiện công thức toán)___

```python
def render_message(message):

    parts = re.split(r'(\$\$.*?\$\$)', message, flags=re.DOTALL)

    for part in parts:

        # latex block
        if part.startswith("$$") and part.endswith("$$"):

            latex_content = part[2:-2].strip()

            st.latex(latex_content)

        # normal text
        else:

            if part.strip():
                st.markdown(part)

```

##Xử lí UI

```python
initial_bot_mess = "Hí chào cậu, tớ là Casio-chan. Tớ có thể giúp cậu bài gì? Tớ chuyên giải Toán - Tin nè!"

def chatbot_ui():
    st.title("Toán gì cơ? cái này nhắm mắt cũng làm được!")
    st.markdown("## Casĩốồô giải tất!")

    # neu chua co lich su tro chuyen
    if 'conversation_log' not in st.session_state:
        st.session_state.conversation_log = [
            {"role": "assistant", "content": initial_bot_mess}
        ]

    # neu da co lich su tro chuyen, hien thi lai lich su do
    for message in st.session_state.conversation_log:
        if message["role"] != "system":
            with st.chat_message(message["role"]):
                st.render_message(message["content"])
      # user input
  prompt = st.text_area(
      "Nhập đề nghen->",
      height=150,
      placeholder="Ví dụ: Tính tọa độ điểm giao của (P): y = -3x^2 và (d): y = 4x + 6"
  )

  uploaded_file = st.file_uploader(
      "📷 Tải ảnh bài toán",
      type=["png", "jpg", "jpeg"]
  )

  if st.button("⚡ Gửi"):

      if not prompt and not uploaded_file:
          st.warning("Nhập đề hoặc tải ảnh lên trước nhé 😅")
          return

      # hien thi prompt nguoi dung
      with st.chat_message("user"):
          st.markdown(prompt)

      # luu vao lich su tro chuyen
      st.session_state.conversation_log.append(
          {"role": "user", "content": prompt}
      )

      # build history
      recent_messages = st.session_state.conversation_log[-10:]

      history = ""

      for msg in recent_messages:

          role = msg["role"]
          content = msg["content"]
          history += f"{role}: {content}\n"

      # generate response
      response = client.models.generate_content(
          model=MODEL_VERSION,
          contents=history,
          config=types.GenerateContentConfig(
              system_instruction=SYS_PROMPT
          )
      )

      bot_reply = response.text

      # hien thi tra loi tu bot
      with st.chat_message("assistant"):
          render_message(bot_reply)

      # luu vao history
      st.session_state.conversation_log.append(
          {
              "role": "assistant",
              "content": bot_reply
          }
      )

      st.rerun()
```

## Run - chạy

```python
if __name__ == "__main__":
    chatbot_ui()
```
